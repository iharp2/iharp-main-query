```python
import xarray as xr

xa = xr.open_dataset('../data/ERA5/2m_temperature-2023-01-01.nc')
min_lat=60
max_lat=80
min_lon=285
max_lon=350
start_datetime="2023-01-01 00:00:00"
end_datetime="2023-01-01 09:00:00"
a = xa.sel(latitude=slice(max_lat,min_lat), longitude=slice(min_lon,max_lon), time=slice(start_datetime,end_datetime))
a
```

```python
a.to_dataframe()
```

```python
a.t2m[0].plot()
```

```python
import geoviews as gv

gv.extension("bokeh")

gv_dataset = gv.Dataset(a)
gv_dataset.to(gv.Image)
```
